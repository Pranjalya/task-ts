```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes = True)

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

from uszipcode import SearchEngine
```

```python
daily = pd.read_csv("beat19-collab-data\\versions\\2020-06-01\\daily-data.csv")
enrollment = pd.read_csv("beat19-collab-data\\versions\\2020-06-01\\enrollment-data.csv")
```

```python
daily.info()
```

```python
total_records = len(daily)
unique_ids = daily._id.nunique()
print(f"number of records: {total_records}" )
print(f"number of unique _ids': {unique_ids}")
```

**Average number of entries per participant:**

```python
avg_time_steps = total_records / unique_ids
print(f"each unique id has an average of{avg_time_steps: .0f} records")
```

**NaN percent per feature:**

```python
percent_nan = daily.isna().sum() * 100 / len(daily)
missing_values = pd.DataFrame({'column_name' : daily.columns,
                              'percent_nan' : percent_nan})
missing_values
```

```python
percent_nan_1 = daily[daily['version'] == 1].isna().sum() * 100 / len(daily[daily['version'] == 1])
percent_nan_2 = daily[daily['version'] == 2].isna().sum() * 100 / len(daily[daily['version'] == 2])
missing_values = pd.DataFrame({'column_name' : daily.columns,
                              'percent_nan_1' : percent_nan_1,
                              'percent_nan_2' : percent_nan_2})
missing_values.reset_index()
```

```python
percent_of_total_1 = daily[daily['version'] == 1].notna().sum() * 100 / daily.notna().sum()
percent_of_total_2 = daily[daily['version'] == 2].notna().sum() * 100 / daily.notna().sum()
percent_of_values = pd.DataFrame({'column_name' : daily.columns,
                              'percent_of_total_1' : percent_of_total_1,
                              'percent_of_total_2' : percent_of_total_2})
percent_of_values.reset_index()
```

```python
bool_columns = list(daily.select_dtypes(include='bool').columns)
```

```python
version_1 = pd.DataFrame({'bool': [0,1]})
version_2 = pd.DataFrame({'bool': [0,1]})
for i in bool_columns:
    temp_1 = daily[daily['version'] == 1][i].value_counts().reset_index()
    version_1 =version_1.merge(temp_1.apply(lambda x: x*100/len(daily[daily['version'] == 1]))[[i]], how = 'left', left_on = 'bool', right_index = True )
    temp_2 = daily[daily['version'] == 2][i].value_counts().reset_index()
    version_2 =version_2.merge(temp_2.apply(lambda x: x*100/len(daily[daily['version'] == 2]))[[i]], how = 'left', left_on = 'bool', right_index = True )
```

```python

search = SearchEngine(simple_zipcode=False)

counties = []
for i in enrollment.zip_postcode:
    counties.append(search.by_zipcode(i).county)

enrollment.insert(10, "county", counties )
```

```python
count_enroll = daily[['date','_id']].merge(enrollment[['_id','county']], how = 'left',on = '_id')
count_enroll = count_enroll.groupby(['date','county']).count()
count_enroll = pd.pivot_table(count_enroll, columns = 'date', index = 'county')
count_enroll.columns = count_enroll.columns.droplevel(0)
count_enroll = count_enroll.reset_index().rename_axis(None, axis=1)
```
